```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("predictions_7f_10.csv", header=None).values
```

```python
data_origin = pd.read_csv("jester-data-1.csv", header=None).values
```

```python
values = [zip(data_origin[i], data[i]) for i in range(0,data.shape[0])]
comparison_data = pd.DataFrame(values)
comparison_data_err = []
for x in range(0, comparison_data.shape[1]):
    for y in range(0, comparison_data.shape[0]):
        if (comparison_data[x][y][0] == 99.0):
            comparison_data[x][y] = (np.nan,np.nan)
        else:
            comparison_data_err.append(comparison_data[x][y][0] - comparison_data[x][y][1])
mse = (np.array(comparison_data_err) ** 2).mean()
#comparison_data.columns = data_origin.columns
comparison_data.applymap(lambda x: ["(%2.3f|%2.3f)"%(x[0],x[1]) ])
```
